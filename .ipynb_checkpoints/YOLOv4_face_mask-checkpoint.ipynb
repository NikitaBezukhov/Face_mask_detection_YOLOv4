### Dataset consists of 853 unique images of different sizes and aspect ratios of people with/without face mask and with mask worn incorrectly. (so 3 classes in total)
### Annotations are in Pascal VOC format, so beforehand I converted them to YOLO DarkNet format.
### Even though DarkNet supports letter_box (0 padding image to square while preserving aspect ratios) during training and inference, it doesnt support letter_box for anchor box k-means calculation. So to get good anchors I padded images with bboxes to square beforehand using Albumentations.
### I split data into train/val sets as 85/15%.

```python
# Next we calculate new anchor boxes for our dataset and then we need to write them into out custom 
# yolov4-obj.cfg and yolov4-obj-val-mode.cfg and put these files into darknet/data folder. 
# (cfg files define model architecture, so you should choose the right one from https://github.com/AlexeyAB/darknet repo 
# and change hyperparameters in them depending on your needs)
```

```python
!./darknet detector calc_anchors data/obj.data -num_of_clusters 9 -width 640 -height 640
```

### For hyper parameters i've chosen:
#### batch=64
#### subdivisions=16
#### width=640
#### height=640
#### channels=3
#### momentum=0.949
#### decay=0.0005
#### angle=0
#### saturation = 1.5
#### exposure = 1.5
#### hue=.1
#### flip=1
#### learning_rate=0.001
#### burn_in=1000
#### max_batches = 6000
#### policy=steps
#### steps=2400,4800,5400
#### scales=.1,.1,.1
#### mosaic=1
#### letter_box=1
#### random=0
### From training experiments I found model to start wandering around plateau from 2000-2400 iterations, so weight decay happens on 2400, 4800 and 5400 iterations. Subdivision was chosen as 16 and random network scaling as 0 because of 16gb GPU limitations.

```python
# Now we can train the model
```

```python
# %%capture
!./darknet detector train /content/darknet/data/obj.data /content/darknet/data/yolov4-obj.cfg /content/darknet/yolov4-csp-swish.conv.164 -dont_show -map
```

```python
# Next we can look at the chart of our mAP/Loss over training time.
```

```python
def imShow(path):
  import cv2
  import matplotlib.pyplot as plt
  %matplotlib inline

  image = cv2.imread(path)
  height, width = image.shape[:2]
  resized_image = cv2.resize(image,(3*width, 3*height), interpolation = cv2.INTER_CUBIC)

  fig = plt.gcf()
  fig.set_size_inches(18, 10)
  plt.axis("off")
  plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
  plt.show()
```

```python
imShow('chart.png')
```

## In the end we were able to achieve 86.75 mAP50 which I consider a pretty good result since we only had 725 training images. The lowest AP was on "incorrect mask wearing" class - 77.53%, which is expected, since upon inspection its representation in the dataset was pretty low. With mask and without mask APs were 93.65% and 89.07% respectively which I think is really good for this small of a dataset. And for confidence thresholds of 0.25 and 0.50 average IoU was around 76-78%, which also is a good result.
